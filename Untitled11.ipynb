# **Visualição de dados com Pandas**





```
import pandas as pd
```

```
# Importando csv, 
# Colocando Date no formato certo para data e usando data como index também.

df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/BBAS3.SA.csv")
df.Date = pd.to_datetime(df.Date)
df.set_index('Date', inplace=True)
```

```
df.head()
```

## Gráfico de Linhas

```
# Grafico de High

df.High.plot();
```

```
# Gráfico de Low

df.Low.plot();
```

```
# Gráfico de Volume

df.Volume.plot();
```

## Histograma

```
df.High.hist();
```

```
df.Volume.hist();
```

## Gráfico de Dispersão (*Scatter Plot*)

```
# O Gráfico de Dispersão é um gráfico ideal para vera correlação entre duas variáveis quaisquer.
# Para tanto será usado outro dataset, o consumo de cerveja em uma região universitária em São Paulo.

df_2 = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/consumo_cerveja.csv")

df_2.head()
```

```
df_2.plot.scatter('media','consumo');
```

## Boxplot

```
# Nos Boxplot tem-se os valores maximos e minimos, quartis e mediana 
# e os Outliers que são ponto representados acima ou abaixo dos valores 
# mínimos.

```

```
df_2.consumo.plot(kind="box")
```
